```
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
```

```
# creating moons dataset
data = make_moons(n_samples=10000, noise=0.4, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(data[0], data[1], test_size=0.2, random_state=42)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
```

```
# defining models for voting classifier 
lin_clf = LogisticRegression()
tre_clf = DecisionTreeClassifier()
rnd_clf = RandomForestClassifier()
svm_clf = SVC()

vot_clf = VotingClassifier(estimators=[('lr', lin_clf), 
                                       ('dt', tre_clf),
                                       ('rf', rnd_clf),
                                       ('sc', svm_clf)], voting='hard')
```

```
# accuracy evaluation of each model 
for clf in [lin_clf, tre_clf, rnd_clf, svm_clf, vot_clf]:
  clf.fit(X_train, y_train)
  print(clf.__class__.__name__, accuracy_score(clf.predict(X_test), y_test))
```

```
# defining models for voting classifier 
lin_clf = LogisticRegression()
tre_clf = DecisionTreeClassifier()
rnd_clf = RandomForestClassifier()
svm_clf = SVC(probability=True)

vot_clf = VotingClassifier(estimators=[('lr', lin_clf), 
                                       ('dt', tre_clf),
                                       ('rf', rnd_clf),
                                       ('sc', svm_clf)], voting='soft')
```

```
# accuracy evaluation of each model 
for clf in [lin_clf, tre_clf, rnd_clf, svm_clf, vot_clf]:
  clf.fit(X_train, y_train)
  print(clf.__class__.__name__, accuracy_score(clf.predict(X_test), y_test))
```
